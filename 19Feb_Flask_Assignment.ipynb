#### Q3. What is App routing in Flask? Why do we use app routes?

In Flask, App routing is the process of mapping URLs to specific functions that will handle requests made to those URLs. It involves defining routes for your application, which determines what function is executed when a user visits a particular URL.

Routes are created using the @app.route() decorator in Flask, and each route is associated with a specific view function that will be executed when the route is requested. Here's an example:

    from flask import Flask

    app = Flask(__name__)

    @app.route('/')
    def index():
        return 'Hello, World!'

    @app.route('/about')
    def about():
        return 'This is the about page.'

    if __name__ == '__main__':
        app.run(debug=True)
        
In this example, we've defined two routes: one for the homepage ('/') and one for the about page ('/about'). Each route is associated with a view function (index() and about(), respectively) that returns some HTML content.

We use app routes in Flask to create a well-organized and maintainable web application. By mapping URLs to specific functions, we can create a logical structure for our application that makes it easy to add new functionality or modify existing behavior. App routing also allows us to handle different types of HTTP requests, such as GET or POST requests, which can be useful for creating interactive web applications.

Overall, app routing is an essential part of building web applications in Flask, and it helps to make our applications more modular, flexible, and scalable.

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. This function generates a URL to the specified endpoint with the given method and arguments.

Here's an example of how to use url_for():

    from flask import Flask, url_for

    app = Flask(__name__)

    @app.route('/hello/<name>')
    def hello(name):
        return f'Hello, {name}!'

    with app.test_request_context():
        print(url_for('hello', name='Jasna'))
        
In this example, the url_for() function is used to generate a URL for the hello() function, which takes a name parameter. The generated URL will look something like this: /hello/John.

Note that the url_for() function can be used within a request context or outside of it. When used outside of a request context, as in the example above, you need to create a test request context using the test_request_context() method to simulate a request context.
