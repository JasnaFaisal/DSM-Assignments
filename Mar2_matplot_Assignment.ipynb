#### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular data visualization library in Python that allows you to create a wide range of static, animated, and interactive visualizations. It is used for creating high-quality 2D and 3D plots, charts, and graphs in Python.

Matplotlib is used because it offers a comprehensive range of features for data visualization, and it is highly customizable. It provides a vast array of tools for creating various types of plots, including line plots, scatter plots, bar plots, histograms, pie charts, and more. It also allows you to customize the appearance of your plots, such as adjusting colors, fonts, labels, axes, and legends, among others.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

    1.Line Plot: A line plot is a type of plot that displays data as a series of points connected by straight lines. It is used to show the trend or pattern in data over time or across different variables.

    2.Scatter Plot: A scatter plot is a type of plot that displays the relationship between two variables. It is used to show how much one variable is affected by another or to detect outliers in the data.

    3.Bar Plot: A bar plot is a type of plot that displays data as rectangular bars. It is used to compare the values of different variables or to show how a variable changes over time.

    4.Histogram: A histogram is a type of plot that displays the distribution of a variable. It is used to show how often a certain range of values occurs in the data.

    5.Pie Chart: A pie chart is a type of plot that displays data as a circular chart divided into slices. It is used to show how much each category contributes to the total value of the data.

#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of data visualization that displays the relationship between two continuous variables. It is created by plotting one variable on the x-axis and another variable on the y-axis, and then representing each data point as a dot or point on the graph.

The scatter plot is used to visually examine the correlation or relationship between two variables. It helps in determining whether the two variables are related and how strong the relationship is. If there is a positive correlation between the two variables, the data points will generally fall along an upward-sloping line. If there is a negative correlation, the data points will generally fall along a downward-sloping line. If there is no correlation, the data points will be scattered randomly across the graph.

Scatter plots are commonly used in many fields, including social sciences, engineering, business, and finance, among others. They are useful in identifying patterns, trends, outliers, and clusters in the data, as well as in making predictions and drawing conclusions about the data. They can also be used to compare the results of different experiments or to visualize the results of a regression analysis.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.title('Graph of X vs Y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.scatter(x,y)
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

The subplot() function is used in Matplotlib to create multiple plots in a single figure. It allows you to display different plots side by side or stacked on top of each other, making it easier to compare and analyze multiple data sets.

The subplot() function takes three arguments: the number of rows of plots, the number of columns of plots, and the index of the plot you want to create. For example, subplot(2, 2, 1) creates a figure with 2 rows and 2 columns of plots, and specifies that the first plot should be created.

The subplot() function is particularly useful when you want to create multiple plots that share the same axes or have different scales. By using this function, you can adjust the size and location of each plot in the figure, and customize the appearance of each plot separately.

Additionally, the subplot() function allows you to create complex layouts that contain multiple plots of different sizes and orientations. This can be useful when you want to create a customized data visualization that presents a lot of information in a compact and easy-to-understand format.

In summary, the subplot() function is used to create multiple plots in a single figure, which helps in comparing and analyzing multiple data sets, adjusting the size and location of each plot, and creating complex layouts that present information in a compact and easy-to-understand format.







```python
#For line 1: 
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
#For line 2: 
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
#For line 3: 
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
#For line 4: 
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
# Create a figure with 2 rows and 2 columns of plots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot the first line plot in the first subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Plot 1')

# Plot the second line plot in the second subplot
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Plot 2')

# Plot the third line plot in the third subplot
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Plot 3')

# Plot the fourth line plot in the fourth subplot
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Plot 4')

# Add a suptitle and adjust the layout
fig.suptitle('Four Line Plots', fontsize=20)
fig.tight_layout()

# Show the plot
plt.show()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot is a type of data visualization that displays data using rectangular bars of equal width, where the height or length of each bar represents the value of a particular variable. Bar plots can be horizontal or vertical, and are commonly used to compare categorical data or discrete variables.

A bar plot is used to represent data that can be broken down into categories or discrete values, such as counts, frequencies, proportions, or percentages. It is useful for showing the distribution of data and identifying patterns or trends within the data. Bar plots are particularly effective for visualizing data that is not continuous, but rather consists of discrete categories or groups.

Bar plots are often used in many fields, including social sciences, marketing, business, and finance, among others. They can be used to compare data across different categories or groups, to show changes in data over time, or to compare the relative sizes of different data sets. Bar plots are also useful in identifying outliers or extreme values within the data, and in detecting any anomalies or discrepancies that may need further investigation.

In summary, a bar plot is a type of data visualization that displays data using rectangular bars of equal width, where the height or length of each bar represents the value of a particular variable. It is used to represent data that can be broken down into categories or discrete values, and is particularly useful for comparing data across different categories or groups, identifying patterns or trends within the data, and detecting outliers or anomalies.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.figure(figsize=(8,6))
plt.title('Plot of Sales Profit')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.bar(company,profit)
plt.grid()
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot is a type of data visualization that displays the distribution of a continuous variable using a box-and-whisker plot. The plot is composed of a box that represents the interquartile range (IQR) of the data, and whiskers that extend from the box to represent the range of the data that falls within a certain distance from the median. Box plots can also include outliers, which are plotted as individual points outside the whiskers.

A box plot is used to summarize the distribution of a variable and to identify any outliers or extreme values within the data. It provides a quick and easy way to compare the distribution of a variable across different groups or categories. For example, a box plot can be used to compare the distribution of test scores for students in different classes, or to compare the distribution of salaries for employees in different departments.

Box plots are particularly useful in identifying the presence of outliers or extreme values within the data. Outliers can be indicative of errors in the data or of unusual circumstances that need further investigation. Box plots can also help in identifying skewness or asymmetry in the distribution of the data, as well as in detecting any patterns or trends within the data.

In summary, a box plot is a type of data visualization that displays the distribution of a continuous variable using a box-and-whisker plot. It is used to summarize the distribution of a variable, to identify outliers or extreme values within the data, and to compare the distribution of a variable across different groups or categories. Box plots are particularly useful in detecting patterns or trends within the data, and in identifying any anomalies or discrepancies that may need further investigation.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
```
